```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
image1 = cv2.imread("image.jpg")
image2 = cv2.cvtColor(image1,cv2.COLOR_BGR2RGB)
kernel = np.ones((11,11),np.float32)/121
image3 = cv2.filter2D(image2,-1,kernel)
plt.figure(figsize = (9,9))
plt.subplot(1,2,1)
plt.imshow(image2)
plt.title("Original")
plt.axis("Off")
plt.subplot(1,2,2)
plt.imshow(image3)
plt.title("Filtered")
plt.axis('off')
```
